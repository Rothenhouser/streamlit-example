```python
import pandas as pd
from data import _get_station_info, get_temperatures, _get_useful_stations_from_data_urls

all_stations = _get_station_info()
all_stations
```

```python
useful_stations = _get_useful_stations_from_data_urls()
useful_stations
```

```python
useful_stations.join(
    all_stations.set_index('Stations_id'),
    on='station_id'
    ).reset_index(drop=True)
```

```python
stations = pd.merge(
    all_stations, 
    useful_stations, 
    how='inner', 
    left_on='Stations_id', 
    right_on='station_id',
    validate='one_to_one'
    )

stations
```

```python
max_temps = get_temperatures(stations['url'].iloc[1])
max_temps
```

```python
max_temps[max_temps['MX_TX'] == -999] = pd.NA
max_temps
```
